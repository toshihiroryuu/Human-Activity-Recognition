### Sensor Data and Time Stamp

```python
import ast

i = 0
x = []
acc_x, acc_y, acc_z = [], [], []
gyro_x, gyro_y, gyro_z = [], [], []
for msg in consumer1:
    byte_str = msg[6]
    dict_str = byte_str.decode("UTF-8")
    data = ast.literal_eval(dict_str)

    if "gyroscope" in data:
        print("Time Stamp : ", data["@timestamp"])
        print("Accelerometer : ", data["accelerometer"])
        print("Gyroscope : ", data["gyroscope"])
        
        x.append(i)
        acc_x.append(data["accelerometer"][0])
        acc_y.append(data["accelerometer"][1])
        acc_z.append(data["accelerometer"][2])
        
        ax1.plot(x, acc_x, color='b')
        ax2.plot(x, acc_y, color='g')
        ax3.plot(x, acc_z, color='r')
        
#         gyro_x.append(data["gyroscope"][0])
#         gyro_y.append(data["gyroscope"][1])
#         gyro_z.append(data["gyroscope"][2])
        
#         ax4.plot(x, gyro_x, color='b')
#         ax5.plot(x, gyro_y, color='g')
#         ax6.plot(x, gyro_z, color='r')
        
        fig.canvas.draw()
    
#         ax1.set_xlim(left=max(0, i-50), right=i+50)
        ax2.set_xlim(left=max(0, i-50), right=i+50)
#         ax2.set_xlim(left=max(0, i-50), right=i+50)
    
        time.sleep(0.1)
        i += 1
```
